```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("high_risk_summary.csv", names=["Age", "Count"])


plt.figure(figsize = (11, 6))
plt.bar(df["Age"], df["Count"], color="tomato")
plt.title("High BMI & Diabetic patients by Age")
plt.xlabel("Age")
plt.ylabel("Number of High-risk Individuals")
plt.grid(True)
plt.tight_layout()
plt.show()
```
