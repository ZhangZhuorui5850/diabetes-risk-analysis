```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 上传好后读取文件
df = pd.read_csv("diabetes.csv" )

# 替换一些明显无效的值为 NaN（方便后续可视化不受污染）
df['Glucose'] = df['Glucose'].replace(0, pd.NA)
df['BMI'] = df['BMI'].replace(0, pd.NA)
df['BloodPressure'] = df['BloodPressure'].replace(0, pd.NA)
df['Insulin'] = df['Insulin'].replace(0, pd.NA)
```

```python
# 计算分组平均
bmi_avg = df.groupby("Outcome")["BMI"].mean()

# 画柱状图
plt.figure(figsize=(6,4))
bmi_avg.plot(kind="bar", color=["green", "red"])
plt.title("Average BMI by Diabetes Outcome")
plt.xticks([0,1], ["No Diabetes", "Diabetes"])
plt.ylabel("BMI")
plt.tight_layout()
plt.show()
```

```python
age_avg = df.groupby("Outcome")["Age"].mean()

plt.figure(figsize=(6,4))
age_avg.plot(kind="bar", color=["blue", "orange"])
plt.title("Average Age by Diabetes Outcome")
plt.xticks([0,1], ["No Diabetes", "Diabetes"])
plt.ylabel("Age")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df[df["Outcome"]==0]["Glucose"].dropna(), label="No Diabetes", fill=True)
sns.kdeplot(data=df[df["Outcome"]==1]["Glucose"].dropna(), label="Diabetes", fill=True, color="red")
plt.title("Glucose Distribution by Outcome")
plt.xlabel("Glucose")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
